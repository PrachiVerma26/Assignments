```python
# Import required libraries
import pandas as pd

# 1. Load into pandas : Create student performance dataset
students = {
    "Name": ["Adarsh", "Shreya", "Yash", "Aman"],
    "Marks": [89, 58, 75, 95],
    "Hours Studied": [4, 1, 3, 6]
}

df = pd.DataFrame(students)
df
```

```python
# 2. Add Performance column based on marks obtained( Pass if Marks > 65 else Fail)
df["Performance"] = df["Marks"].apply(lambda marks: "Pass" if marks > 65 else "Fail")

print("Student performance analysis:")
df
```

```python
# Import Matplotlib for visualization
import matplotlib.pyplot as plt

# Create line chart showing relationship between study hours and marks
plt.plot(df["Hours Studied"], df["Marks"], marker="o")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Hours Studied vs Marks")

plt.show()
```

```python
# Create scatter plot for study hours and marks
plt.scatter(df["Hours Studied"], df["Marks"])

plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

plt.show()
```

```python
# Import Seaborn library
import seaborn as sns

# Create bar plot comparing performance category and marks
sns.barplot(data=df, x="Performance", y="Marks", hue= "Performance", palette = ["lightgreen", "red"])

plt.title("Performance vs Marks")
plt.show()
```
