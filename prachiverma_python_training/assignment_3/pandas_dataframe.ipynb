```python
# 1. Import Pandas library
import pandas as pd
```

```python
# Create employee dataset using a dictionary
employees = {
    "Name": ["Rahul", "Priya", "Amit", "Anuj"],
    "Age": [25, 30, 28, 35],
    "Department": ["HR", "IT", "Finance", "IT"],
    "Salary": [30000, 50000, 45000, 60000]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(employees)

# Display complete DataFrame
print("Employee details: ")
df
```

```python
# 1. Display first two employee records
print("First two employee records:")
df.head(2)
```

```python
# 2. Generate summary statistics for numerical columns
print("Summary Statistics: ")
df.describe()
```

```python
# 3. Filter employees belonging to IT department
it_employees = df[df["Department"] == "IT"]

print("Filtered employee records for the IT department:")
it_employees
```

```python
# 4. Add Bonus column (10% of salary)
df["Bonus"] = df["Salary"] * 0.10

# Display updated DataFrame
print("Employee records with calculated bonus:")
df
```
