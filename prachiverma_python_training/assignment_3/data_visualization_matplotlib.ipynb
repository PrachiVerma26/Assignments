```python
# Import Matplotlib library
import matplotlib.pyplot as plt

# Create dataset
departments = ["IT", "Sales and Marketing", "Finance", "HR"]
employees = [15, 5, 7, 8]
```

```python
# Create bar chart showing employee count by department
plt.bar(departments, employees)

plt.title("Employees by Department")
plt.xlabel("Department")
plt.ylabel("Employee Count")

plt.show()
```

```python
# Create line chart
plt.plot(departments, employees, marker="o")

plt.title("Department Employee Trend")
plt.xlabel("Department")
plt.ylabel("Employee Count")

plt.show()
```

```python
# Create sample employee salary data for histogram analysis
salaries = [30000, 50000, 45000, 60000, 25000, 90000, 35000]

# Plot histogram
plt.hist(salaries)

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

```python
# Create age and salary dataset for scatter plot
ages = [22, 25, 28, 31, 35, 40, 45]
salaries = [45000, 35000, 14000, 55000, 65000, 80000, 55000]

# Create scatter plot
plt.scatter(ages, salaries)

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")

plt.show()
```
