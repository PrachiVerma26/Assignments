```python
# Import Pandas library
import pandas as pd
```

```python
# Create employee dataset
employees = {
    "Name": ["Garima", "Jasmine", "Amit", "Reyansh", "Missi", "Shaurya", "Zayn"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance", "HR"],
    "Salary": [30000, 50000, 45000, 60000, 25000, 90000, 35000]
}

# Convert the employee dataset into a Pandas DataFrame
df = pd.DataFrame(employees)

print("Employee dataset:")
df
```

```python
# 1. Calculate average salary department-wise

# Use groupby() and mean() to calculate the average salary for each department
average_salary = df.groupby("Department")["Salary"].mean().reset_index(name="Average Salary")

print("Average salary for each department:")
average_salary
```

```python
# 2. Find max salary by department

# Use groupby() and max() to find the highest salary in each department
max_salary = df.groupby("Department") ["Salary"].max().reset_index(name="Maximum Salary")

print("Highest salary in each department:")
max_salary
```

```python
# 3. Count employees in each department

# Use groupby() and count() to calculate the number of employees in each department
employee_count = df.groupby("Department")["Name"].count().reset_index(name="Employee Count")

print("Number of employees in each department:")
employee_count
```
