```python
# Import required libraries
import pandas as pd
import numpy as np
```

```python
# Create a sample dataset with missing values
data = {
    "Name": ["Rahul", "Priya", "Anuj"],
    "Age": [25, np.nan, 29],
    "Salary": [30000, 40000, np.nan]
    }
df = pd.DataFrame(data)
df
```

```python
# 1. detect missing values in the dataset

print("Missing value analysis using isna():")
# Use isna() to identify missing (NaN) values in each cell
display(df.isna())

print("Total Missing values in each column:")
# Use isna().sum() to count missing values in each column 
df.isna().sum()
```

```python
# 2. Replace missing Age with mean
mean_age = df["Age"].mean()
print(f"Mean Age: {mean_age}")

# Use fillna() to fill NaN values in the Age column with the calculated mean age
df["Age"] = df["Age"].fillna(mean_age)
df["Age"] = df["Age"].fillna(mean_age)

print("Dataset after filling missing age values:")
df
```

```python
# 3. Replace missing salary values with zero 
df["Salary"] = df["Salary"].fillna(0)

print("Updated dataset after handling missing salary values:")
df
```
