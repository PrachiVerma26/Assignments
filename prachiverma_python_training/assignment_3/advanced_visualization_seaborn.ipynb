```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create employee dataset
df = pd.DataFrame({
    "Name": ["Adarsh", "Shreya", "Yash", "Aman"],
    "Age": [25, 30, 28, 35],
    "Department": ["Sales", "IT", "HR", "Marketing"],
    "Salary": [35000, 60000, 75000, 50000]
})

df
```

```python
# 1. Create bar plot for department and salary
sns.barplot(
    data=df,
    x="Department",
    y="Salary",
    hue="Department",   # Assign different colors to each department
    palette=["skyblue", "salmon", "lightgreen", "yellow"]
)

plt.title("Department vs Salary")
plt.show()
```

```python
# 2. Create box plot to visualize salary distribution
sns.boxplot(x=df["Salary"], color = "Lightgreen")

plt.title("Salary Distribution of Employees")
plt.show()
```

```python
# Select numerical columns for correlation analysis
numeric_df = df[["Age", "Salary"]]

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Visualize correlation values using a heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,      # Display correlation values in each cell
    cmap="Blues"     # Apply a blue color scheme
)

plt.title("Correlation Between Age and Salary")

plt.show()
```
